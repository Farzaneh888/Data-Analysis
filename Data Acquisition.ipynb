```python
import pandas as pd
import numpy as np
```

```python

# Reading the raw data
url_raw = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

# Defining column headers (taken from the dataset documentation)
columns = [
    "symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"
]

# Creating a DataFrame
df = pd.read_csv(url_raw, header=None)  # Load the raw data without headers
df.columns = columns  # Assign the column headers to the DataFrame

# Saving the data as a CSV file
file_path = 'auto.csv'


print(f'Data has been saved to {file_path}')
```

```python
file_name="auto.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
#clean data
df1=df.replace('?',np.NaN)
df1.info()
```

```python
df1.isnull().sum()
```

```python
# Strip whitespace from column names to ensure proper matching
df1.columns = df1.columns.str.strip()

# Convert any non-numeric values in selected columns to NaN (including extra spaces or symbols)
df1[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']] = \
    df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']].apply(pd.to_numeric, errors='coerce')

# Recheck for missing values after conversion and filling
print("Before filling missing values:")
print(df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']].isna().sum())

# Calculate the mean of each column (ignoring NaN values)
mean_values = df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']].mean()

# Replace NaN values with the calculated mean for each column
df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']] = \
    df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']].fillna(mean_values)

# Recheck for missing values after filling
print("After filling missing values:")
print(df1[['normalized-losses', 'bore', 'stroke','horsepower', 'peak-rpm']].isna().sum())
```
